```python
from _pipeline import create_payload, model_req

# Define the message related to Python lists
MESSAGE = "How does list comprehension work in Python?"

# Constructing the Multi-Turn Prompt
PROMPT_1 = f"""
Explain "{MESSAGE}" using *simple examples*.  
1 Explain what list comprehension is.  
2 Give a basic example.  
3 Wait for user input before continuing.
"""

# Creating the payload for step 1
payload1 = create_payload(target="ollama",
                          model="llama3.2:latest",
                          prompt=PROMPT_1,
                          temperature=0.7, 
                          num_ctx=200, 
                          num_predict=200)

# Send request to the model
time1, response1 = model_req(payload=payload1)
print("Step 1 Response:", response1)

# Constructing the second step prompt based on user input
PROMPT_2 = f"""
Now, based on the previous explanation, provide a *real-world use case* for list comprehensions.
Example: Filtering even numbers from a list.
"""

# Creating the payload for step 2
payload2 = create_payload(target="ollama",
                          model="llama3.2:latest",
                          prompt=PROMPT_2,
                          temperature=0.7, 
                          num_ctx=200, 
                          num_predict=200)

# Send request to the model
time2, response2 = model_req(payload=payload2)
print("Step 2 Response:", response2)

if time2: print(f'Time taken: {time2}s')
```
