```python
from _pipeline import create_payload, model_req

# Define the message related to Excel
MESSAGE = "How does VLOOKUP work in Excel with a practical example?"

# Constructing the Role-Based Prompt
PROMPT = f"""
You are an *Excel trainer* helping students understand Excel functions.  
Your task is to explain *VLOOKUP* in a *step-by-step manner* using a *real-world example*.  
Make sure to:
1. Explain what VLOOKUP does.  
2. Provide a simple dataset example.  
3. Show an actual VLOOKUP formula.  
4. Explain how the lookup process works.  

Now, answer this query: "{MESSAGE}"
"""

# Creating the payload for the AI model request
payload = create_payload(target="ollama",
                         model="llama3.2:latest",
                         prompt=PROMPT,
                         temperature=0.7, 
                         num_ctx=150, 
                         num_predict=250)

# Send request to the model
time, response = model_req(payload=payload)

# Print the response
print(response)
if time: print(f'Time taken: {time}s')
```
