```python
from _pipeline import create_payload, model_req

# Define the message related to SQL
MESSAGE = "Why is my SQL query running slow?"

# Constructing the Tree-of-Thought Prompt
PROMPT = f"""
Analyze the possible reasons for slow SQL query performance.  
1 Generate at least *three possible causes* for slowness.  
2 For each cause, provide a *detailed explanation*.  
3 Rank the causes based on likelihood and impact.  
4 Suggest *best practices* for optimization.
"""

# Creating the payload for the AI model request
payload = create_payload(target="ollama",
                         model="llama3.2:latest",
                         prompt=PROMPT,
                         temperature=0.7, 
                         num_ctx=300, 
                         num_predict=400)

# Send request to the model
time, response = model_req(payload=payload)

# Print the response
print(response)
if time: print(f'Time taken: {time}s')
```
